```python
import pandas as pd
reviews = pd.read_csv("C:/Users/aliemir/Desktop/yazilim/Python/kaggle/Padas/Datafiles/winemag-data-130k-v2.csv", index_col = 0)
pd.set_option('max_rows', 5)
```

```python
reviews
```

```python
reviews.groupby("points").points.count() #count of each point which appliead to wines.
```

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0]) #first review from every wine reviews
```

```python
reviews.groupby(["country", "province"]).apply(lambda df: df.loc[df.points.idxmax()]) #best wine of provinces
```

```python
countries_reviewed = reviews.groupby(['country', 'province']).price.agg([len, min, max]) 
countries_reviewed #Amount of wines, maximum price of wines and minimum price of wines from each province
```

```python
countries_reviewed.sort_values(by='len') #sorting reviews according to amount of wines from each province
```

```python
countries_reviewed.sort_values(by = "len", ascending = False) #sorting reviews according to amount of wines from each province in a descending order
```
